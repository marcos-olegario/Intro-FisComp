```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
quadrante1 = np.loadtxt('traj1.dat')
quadrante2 = np.loadtxt('traj2.dat')
quadrante3 = np.loadtxt('traj3.dat')
quadrante4 = np.loadtxt('traj4.dat')

plt.plot(quadrante1[:,0], quadrante1[:,1], label="1º Quarto", lw=2)
plt.plot(quadrante2[:,0], quadrante2[:,1], label="2º Quarto", lw=2)
plt.plot(quadrante3[:,0], quadrante3[:,1], label="3º Quarto", lw=2)
plt.plot(quadrante4[:,0], quadrante4[:,1], label="4º Quarto", lw=2)

plt.plot([0], [0], 'o', label = 'Sol', linewidth = 15, color = '#FFC000')

plt.legend()
plt.axis('equal')
plt.xlabel('$x$ (UA)')
plt.ylabel('$y$ (UA)')
plt.title('Gráfico A - Órbita de Mercúrio')

plt.savefig('grafA.pdf', format='pdf', dpi=300)

plt.show(True)
```

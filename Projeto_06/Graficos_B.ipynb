```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
orbita1 = np.loadtxt('trajB1_out_fm1.dat')

plt.plot(orbita1[:,1], orbita1[:,2], label="Órbita da Terra", lw=1)
plt.plot(orbita1[:,3], orbita1[:,4], label="Órbita de Júpiter", lw=1)

plt.plot([0], [0], 'o', label = 'Sol', linewidth = 15, color = '#FFC000')

plt.legend()
plt.axis('equal')
#plt.xlim([-10, 10])
#plt.ylim([-8, 8])
plt.xlabel('$x$ (UA)')
plt.ylabel('$y$ (UA)')
plt.title('Simulação de 3-corpos: Massa de Júpiter = $M_j$')

#plt.text(-7, 5, 'Massa de Jupiter = $M_j$', fontsize=10)
plt.savefig('grafB1.pdf', format='pdf', dpi=300)

plt.show(True)
```
